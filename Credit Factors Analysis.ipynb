```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the dataset
import os
import pandas as pd
outname = 'credit_score.csv'

outdir = 'dir'
if not os.path.exists(outdir):
    os.mkdir(outdir)

credit_score = pd.read_csv(outname)

 
# Create the dataframe
credit_df = pd.DataFrame(credit_score)

print(credit_df.head())
```
