# Identify outliers in credit_score.csv by excel column (T - AN) compared to AO(Credit_Score)

```python
#FUNCTION TO GET OUTLIERS FROM DF
def get_outliers (df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    difference = Q3-Q1
    lower = Q1-1.5* difference
    upper = Q3+1.5* difference
    outliers = (df<lower)|(df>upper)
    return outliers
```

```python
#get names from problematic data from T-AN
columns_t_an = ['YEARLY_GROCERIES_TO_YEARLY_INCOME_RATIO',
           'YEARLY_GROCERIES_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_GROCERIES_TO-CURRENT_DEBT_RATIO',
           'YEARLY_HEALTH_TO_YEARLY_INCOME_RATIO',
           'YEARLY_HEALTH__TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_HEALTH_TO_CURRENT_DEBT_RATIO',
           'YEARLY_HOUSING_TO_YEARLY_INCOME_RATIO',
           'YEARLY_HOUSING_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_HOUSING_TO_CURRENT_DEBT_RATIO',
           'YEARLY_TAX_TO_YEARLY_INCOME_RATIO',
           'YEARLY_TAX_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_TAX_TO_CURRENT_DEBT_RATIO',
           'YEARLY_TRAVEL_TO_YEARLY_INCOME_RATIO',
           'YEARLY_TRAVEL_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_TRAVEL_TO_CURRENT_DEBT_RATIO',
           'YEARLY_UTILITIES_TO_YEARLY_INCOME_RATIO',
           'YEARLY_UTILITIES_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_UTILITIES_TO_CURRENT_DEBT_RATIO',
           'YEARLY_EXPENDITURE_TO_YEARLY_INCOME_RATIO',
           'YEARLY_EXPENDITURE_TO_CURRENT_SAVINGS_RATIO',
           'YEARLY_EXPENDITURE_TO_CURRENT_DEBT_RATIO'
]

compared_df = credit_score[columns_t_an].sub(credit_score['CREDIT_SCORE'], axis=0)
outliers = compared_df.apply(get_outliers)
```

```python
# columns with outliers relative to credit score and the # of outliers in each column
count_outliers = outliers.sum()
count_outliers
```
