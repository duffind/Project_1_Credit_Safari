# Identify outliers in credit_score.csv by excel column (T - AN) compared to AO(Credit_Score)

```python
#FUNCTION TO GET OUTLIERS FROM DF
def get_outliers (df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    difference = Q3-Q1
    lower = Q1-1.5* difference
    upper = Q3+1.5* difference
    outliers = (df<lower)|(df>upper)
    return outliers
```

```python
#get names from problematic data from T-AN
columns_t_an = ['Yearly Groceries : Yearly Income',
                'Yearly Groceries : Current Savings',
                'Yearly Groceries : Current Debt',
                'Yearly Health : Yearly Income', 
                'Yearly Health : Current Savings',
                'Yearly Health : Current Debt', 
                'Yearly Housing : Yearly Income',
                'Yearly Housing : Current Savings', 
                'Yearly Housing : Current Debt',
                'Yearly Tax : Yearly Income', 
                'Yearly Tax : Current Savings',
                'Yearly Tax : Current Debt', 
                'Yearly Travel : Yearly Income',
                'Yearly Travel : Current Savings', 
                'Yearly Travel : Current Debt',
                'Yearly Utilities : Yearly Income',
                'Yearly Utilities : Current Savings', 
                'Yearly Utilities : Current Debt',
                'Yearly Expenditure : Yearly Income',
                'Yearly Expenditure : Current Savings',
                'Yearly Expenditure : Current Debt'
]

compared_df = credit_score_df[columns_t_an].sub(credit_score_df['Credit Score'], axis=0)
outliers = compared_df.apply(get_outliers)
```

```python
# columns with outliers relative to credit score and the # of outliers in each column
count_outliers = outliers.sum()
count_outliers
```
